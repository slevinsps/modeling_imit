### Равномерное распределение

```python
a=0
b=10
margin_x=2

uniform_dist = UniformDist(a, b)
plot_continuous(uniform_dist, x_left=a-margin_x, x_right=b+margin_x, function='cdf')
plot_continuous(uniform_dist, x_left=a-margin_x, x_right=b+margin_x, function='pdf')
```

### Распределение Эрланга

Pdf: $f(x; k, \lambda) = \frac{\lambda^k x^{k-1} e^{-\lambda x}}{(k-1)!}$ for $x, \lambda > 0$

Cdf: $F(x; k, \lambda) = 1 - \sum_{n=0}^{k-1} \frac{1}{n!} e^{-\lambda x} (\lambda x)^n$

```python
k=9
lambda_=2.0

x_left=0
x_right=20

erlang_dist = ErlangDist(k, lambda_)
plot_continuous(erlang_dist, x_left=x_left, x_right=x_right, function='cdf')
plot_continuous(erlang_dist, x_left=x_left, x_right=x_right, function='pdf')
```
